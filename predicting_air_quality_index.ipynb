```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset here - update the path to your actual CSV file path
data_path = r"C:\Users\tejas\Downloads\data_date.csv"  # or use "C:/Users/tejas/Downloads/data_date.csv"
data = pd.read_csv(data_path)

# Basic data info
print("Data Info:")
print(data.info())

print("\nFirst 5 rows:")
print(data.head())

print("\nSummary Statistics:")
print(data.describe())

# Check for missing values
print("\nMissing values per column:")
print(data.isnull().sum())

# Visualize distribution of AQI
plt.figure(figsize=(8,6))
sns.histplot(data['AQI Value'], bins=50, kde=True)  # Use 'AQI Value' instead of 'AQI'
plt.title('Distribution of AQI')
plt.xlabel('AQI Value')
plt.ylabel('Frequency')
plt.show()

# Since your dataset does not have pollutant columns, we will skip this part for now.
# If you have other relevant columns, you can visualize them similarly.
```

```python
# Data Cleaning
# Convert 'Date' to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Encode 'Status' into numerical values (optional)
data['Status'] = data['Status'].astype('category').cat.codes

# Check the updated DataFrame
print(data.head())

# Visualize AQI over time
plt.figure(figsize=(12, 6))
data.groupby('Date')['AQI Value'].mean().plot()
plt.title('Average AQI Value Over Time')
plt.xlabel('Date')
plt.ylabel('Average AQI Value')
plt.xticks(rotation=45)
plt.show()

# Visualize AQI by Country
plt.figure(figsize=(12, 6))
country_aqi = data.groupby('Country')['AQI Value'].mean().sort_values(ascending=False)
country_aqi.plot(kind='bar')
plt.title('Average AQI Value by Country')
plt.xlabel('Country')
plt.ylabel('Average AQI Value')
plt.xticks(rotation=90)
plt.show()
```
